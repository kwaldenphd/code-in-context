# While Statements

Another way to express `if-then-else` logic in Python is using a `while` statement.

A `while` statement tests for an initial condition, and the lines of code indented under `while` run only when the initial condition is `True`.

## Pseudocode Example

Let's go back to our Hagerty example. Imagine the initial condition is: `Do I have coffee?`

```python
while I DO NOT HAVE COFFEE: # while statement
  if hagerty works: # initial condition
    get coffee from hagerty
  elif charron works: #second condition:
    get coffee from charron
  elif au bon pain works: # third condition
    get coffee from abp
  elif starbucks: # fourth condition
    get coffee from starbucks
```

In this example, our program will continue to test conditions while the initial condition is `TRUE`.

## Python Example

Let's put that in Python syntax.

```python
# while statement with initial condition
while x != y:
  # if statement
  if x > y:
      print("The value of x is greater than the value of y")

      # else if statement
  elif x == y:
      print("The value of x is equal to the value of y")

  # else statement
  else:
      print("The value of x is less than the value of y")
```

As we see in this example, we can nest multiple `if-then-else` logical statements in the same piece of code.
